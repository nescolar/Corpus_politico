```python
import numpy as np
import pandas as pd
import re

from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 1. Corpus
documents = [
    "Political institutions shape public policy and influence democratic governance.",
    "Public policy decisions affect economic development and social outcomes.",
    "Democratic governance relies on institutions, political participation, and public trust."
]

# 2. Vectorizador
vectorizer = CountVectorizer(stop_words="english")

# 3. Ajuste y transformación
X = vectorizer.fit_transform(documents)

# 4. Mostrar vocabulario ordenado
print("Vocabulario:")
vocab = dict(sorted(vectorizer.vocabulary_.items(), key=lambda x: x[1]))
print(vocab)

# 5. Mostrar matriz documento-término
print("Matriz (documentos × términos):")
print(X.toarray())

# 6. Mostrar términos en orden de columnas
print("Términos (columnas):")
print(vectorizer.get_feature_names_out())
```
